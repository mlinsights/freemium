```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

x_n = np.arange(-3,3,step=0.01)
pdf_n = st.norm.pdf(x_n,loc=0,scale=1)
cdf_n = st.norm.cdf(x_n,loc=0,scale=1)

plt.figure()
plt.plot(x_n,pdf_n,color="blue")
plt.title('Normal distribution')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.show()

plt.figure()
plt.plot(x_n,cdf_n,color="blue")
plt.title('Normal distribution: CDF(x)')
plt.xlabel('x')
plt.ylabel('cdf(x)')
plt.show()

```
